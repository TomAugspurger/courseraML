```python
d = loadmat('../files/mlclass-ex3-005/mlclass-ex3/ex3data1.mat')
x, y = d['X'], d['y']

idx = np.random.randint(0, x.shape[0], size=100)
X = x[idx]
```

```python
def display_data(X):
    img = np.empty(shape=(200, 200))
    for i, v in enumerate(X):
        values = v.reshape(20, 20).T
        dx = (i % 10) * 20
        dy = (i // 10) * 20
        img[dx:dx+20, dy:dy+20] = values

    fig, ax = plt.subplots()
    ax.grid(False)
    ax.imshow(img)

display_data(X)
```

### Vectorizing Logistic Regression

```python
# Already did this in exercise 2

def h(theta, x):
    return sigmoid(x.dot(theta))

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def regularized_cost(theta, x, y, h, lambda_=0):
    """
    Cost function for regularized logistic regression.
    
    When lambda_=0 then it's your normal logistic regression.
    """
    m = x.shape[0]
    yhat = h(theta, x).reshape(-1, 1)
    e = -y * np.log(yhat) - (1 - y) * np.log(1 - yhat)
    penalty = (lambda_ / m) * (theta[1:]**2).sum()
    return e.mean() - penalty

def gradient(theta, x, y, h, lambda_=0):
    m = x.shape[0]
    yhat = h(theta, x)
    e = (yhat - y) * x
    (1 / m) * e.sum(0) + (lambda_ / m) * theta
```
